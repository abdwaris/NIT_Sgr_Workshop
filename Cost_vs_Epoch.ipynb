# âœ… Python Code: Visualizing Loss vs Epochs in ANN

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.neural_network import MLPRegressor
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split

# Step 1: Load the data
df = pd.read_excel("Collected Data_PS_GitHub upload.xlsx")
df = df.iloc[:, 1:7]  # Columns B to G
df.columns = ['Slope_Angle', 'CbygammaH', 'phi', 'kh', 'kvr', 'fmp']

X = df.drop(columns='fmp')
y = df['fmp']

# Step 2: Scale the input features
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# Step 3: Split the dataset
X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, test_size=0.2, random_state=42)

# Step 4: Initialize the MLPRegressor for custom training loop
mlp = MLPRegressor(hidden_layer_sizes=(8,), max_iter=1, warm_start=True,
                   solver='sgd', learning_rate_init=0.05, random_state=42)

n_epochs = 100
loss_values = []

# Step 5: Train the model one epoch at a time
for epoch in range(n_epochs):
    mlp.fit(X_train, y_train)
    loss_values.append(mlp.loss_)


```

```python
# Step 6: Plot the loss vs epochs
plt.figure(figsize=(8, 5))
plt.plot(range(1, n_epochs + 1), loss_values, marker='o', linestyle='-')
plt.xlabel('Epoch')
plt.ylabel('Loss (Cost Function)')
plt.title('Loss vs Epochs in ANN')
plt.grid(True)
plt.tight_layout()
plt.show()
```

# ðŸ“Œ Notes for Students:

The loss_ attribute of MLPRegressor shows the cost function value after the most recent training step.

We're using warm_start=True to retain weights between .fit() calls and max_iter=1 so that it trains one epoch at a time.

This simulates a manual control over epochs, allowing you to observe how the network learns gradually.
