# ✅ Goal: Visualize Entropy Curve
We plot entropy as a function of the probability of one class (say, class 1), and show how entropy is highest when p = 0.5.

# ✅ Python Code with Comments & Plot

```python
import numpy as np
import matplotlib.pyplot as plt

# Step 1: Define the entropy function for binary classification
def entropy(p):
    """Binary entropy function"""            #This is a docstring
    return -p * np.log2(p) - (1 - p) * np.log2(1 - p)

# Step 2: Generate a range of probabilities from 0 to 1
p_values = np.linspace(0.001, 0.999, 1000)  # Avoid 0 and 1 to prevent log(0)
entropy_values = entropy(p_values)

# Step 3: Plotting
plt.figure(figsize=(8, 5))
plt.plot(p_values, entropy_values, color='darkblue', linewidth=2)
plt.axvline(0.5, color='red', linestyle='--', label='Max Entropy at p = 0.5')
plt.title('Binary Entropy vs Probability', fontsize=14)
plt.xlabel('Probability of Class 1 (p)', fontsize=12)
plt.ylabel('Entropy H(p)', fontsize=12)
plt.grid(True)
plt.legend()      #Displays a legend on the plot.
plt.tight_layout()  #Automatically adjusts subplot parameters to fit the figure contents nicely — especially when 
                    #labels, titles, or ticks are overlapping or getting cut off.
plt.show()
```

# 📘 Explanation
Entropy quantifies uncertainty.

If p = 0 or p = 1, there's no uncertainty — the class is completely predictable.

If p = 0.5, there's maximum uncertainty — it's like flipping a fair coin.
